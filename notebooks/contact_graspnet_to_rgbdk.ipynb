```python
import numpy as np
import matplotlib.pyplot as plt


file_path = "/home/jacinto/robot-grasp/third_party/contact_graspnet/test_data/0.npy"
data = np.load(file_path, allow_pickle=True)

# Extract the RGB data
rgb_data = data.item().get('rgb')

# Convert BGR to RGB
rgb_data = rgb_data[..., ::-1]

# Display the RGB image
plt.imshow(rgb_data)
plt.axis('off')  # Turn off axis
plt.show()

max_depth_value = data.item().get('depth').max()
print(max_depth_value)

# Modify the depth values
data.item()['depth'] *= 1000
data.item()['rgb'] = rgb_data

# Save the modified data
save_path = "/home/jacinto/robot-grasp/data/rgbdks/contact_graspnet_rgbdk_0.npy"
np.save(save_path, data)
```

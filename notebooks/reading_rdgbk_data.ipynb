```python
import numpy as np
from scipy.ndimage import binary_dilation
import matplotlib.pyplot as plt
affordance_file = '/home/jacinto/robot-grasp/data/mask_references/carton_milk_reference_20241203/affordance.npy'
affordance_data = np.load(affordance_file, allow_pickle=True).item()

# Show the keys and values
for key in affordance_data.keys():
    print(f"Key: {key}")
    print(f"Value: {affordance_data[key]}")

    # Display the masks as images
    masks = affordance_data['masks']
    for i, mask in enumerate(masks):
        plt.figure()
        plt.imshow(mask, cmap='gray')
        plt.title(f'Mask {i}')
        plt.axis('off')  # Hide the axis
        plt.show()

        # Dilate the mask
        dilated_mask = binary_dilation(mask)
        dilated_mask = binary_dilation(dilated_mask)
        dilated_mask = binary_dilation(dilated_mask)
        dilated_mask = binary_dilation(dilated_mask)
        dilated_mask = binary_dilation(dilated_mask)


        # Display the dilated mask
        plt.figure()
        plt.imshow(dilated_mask, cmap='gray')
        plt.title('Dilated Mask')
        plt.axis('off')  # Hide the axis
        plt.show()
```
